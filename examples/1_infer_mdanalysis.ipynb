# Batch Inference Helper

This notebook demonstrates `geom2vec.data.infer_mdanalysis_folder`, a high-level wrapper
that extracts trajectories with MDAnalysis and runs inference only for files that
have not been processed yet. Configure the topology/trajectory paths below and run
each cell in order.


```python
from pathlib import Path

import torch

from geom2vec import create_model_from_checkpoint, infer_checkpoint_config
from geom2vec.data import infer_mdanalysis_folder
```

```python
# Update these paths for your system
topology_file = (
    "/project/dinner/anton_trajs/Chignolin/DESRES-Trajectory_CLN025-0-protein/CLN025-0-protein/chignolin.pdb"
)
trajectory_folder = (
    "/project/dinner/anton_trajs/Chignolin/DESRES-Trajectory_CLN025-0-protein/CLN025-0-protein/"
)
output_dir = Path("/project/dinner/zpengmei/geom2vec_02/examples/_artifacts/chignolin_infer")
output_dir.mkdir(parents=True, exist_ok=True)

stride = 100  # increase for faster prototyping
selection = "prop mass > 1.1"  # focus on heavy atoms
```

```python
checkpoint_path = Path("/project/dinner/zpengmei/geom2vec_02/geom2vec/checkpoints/visnet_l6_h64_rbf64_r75.pth")
device = "cuda" if torch.cuda.is_available() else "cpu"

# config = infer_checkpoint_config(str(checkpoint_path))
# config

rep_model = create_model_from_checkpoint(
    checkpoint_path=str(checkpoint_path),
    device=device,
)
```

```python
summary = infer_mdanalysis_folder(
    model=rep_model,
    topology_file=topology_file,
    trajectory_folder=trajectory_folder,
    output_dir=str(output_dir),
    stride=stride,
    selection=selection,
    batch_size=32,
    reduction="sum",
    overwrite=False,
)
```
